在產生購買行為前，可能會影響消費者購買的因素:
1. 消費者來自哪裡
2. 購買時間/月份
3. 商品內容

## 可以先看看每筆消費的內容：每一次消費買了多少樣東西，或花了多少錢。從買了多少樣東西下手比較簡單

```python
import pandas as pd

orders_item = pd.read_csv('olist_order_items_dataset.csv ', encoding='utf-8')
orders_item.head(20)
```

```python
orders_item_count = orders_item[['order_id','order_item_id']].groupby(['order_id']).count().reset_index()
orders_item_count = orders_item_count[['order_id','order_item_id']].groupby(['order_item_id']).count().reset_index()
orders_item_count = orders_item_count.rename(columns={"order_id": "order_count", "order_item_id":"order_item_count"})
c = orders_item_count['order_count'].sum()
orders_item_count['order_count_percentage']=orders_item_count['order_count']/c
orders_item_count.head()
```

## 觀察每一筆訂單的內容，會發現大部分一筆訂單都只有購買一項物品

## 因為只有1個物品的訂單佔了整個dataset的90%，所以來看購買月份有沒有影響

```python
orders = pd.read_csv('olist_orders_dataset.csv', encoding='utf-8')
orders.head()
```

```python
time_table = pd.merge(orders_item[['order_id','order_item_id']], orders[['order_id','order_purchase_timestamp']], on='order_id')
time_table.head()
```

```python
time_table['order_purchase_timestamp'] = pd.to_datetime(time_table['order_purchase_timestamp'])
time_table = time_table.set_index('order_purchase_timestamp')
time_table.head()
```

```python
time_table.resample('M', convention='end').count()
```

**可以看到2017年的資料最完整，所以接下來只看2017年的分析。2017年有沒有哪一個月份是購買比較多的?**

```python
time_table2 = time_table[time_table.index.year == 2017]
time_table2.resample('M', convention='end').count()
```

**可以看到每個月分購買的數量差不多，除了11月最多、1月最低這點，其他沒有什麼特別**

## 看看購買的時間(幾點)有沒有比較顯著的

```python
time_table3 = time_table2.resample('360T', closed='right').sum()
time_table3
```

從上面的分析，我們幾乎可以下一個簡單的結論：在這家電商下訂的客人，大部分都只會買一樣東西就結帳。這背後可能有幾個原因：
1. 運費很便宜
2. 民族文化，想到什麼就直接上網訂
3. 人民經濟比較不好，無法一次買很多
