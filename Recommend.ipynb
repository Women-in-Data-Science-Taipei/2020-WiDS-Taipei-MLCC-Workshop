```python
import pandas as pd
import numpy as np
```

### Read and Merge Data

```python
reviews = pd.read_csv('olist_order_reviews_dataset.csv', encoding='UTF-8')
reviews.head()
```

```python
review_column_drop = ['review_comment_title', 'review_comment_message', 'review_creation_date', 'review_answer_timestamp']
reviews = reviews.drop(review_column_drop, axis=1)
reviews.head()
```

```python
orders = pd.read_csv('olist_orders_dataset.csv', encoding='UTF-8')
orders.head()
```

```python
order_column_drop = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
orders = orders.drop(order_column_drop, axis = 1)
orders.head()
```

```python
items = pd.read_csv('olist_order_items_dataset.csv', encoding='UTF-8')
items.head()
```

```python
item_column_drop = ['shipping_limit_date', 'seller_id']
items = items.drop(item_column_drop, axis=1)
items.head()
```

```python
reviews_orders = pd.merge(reviews,orders, on='order_id')
reviews_orders_items = pd.merge(reviews_orders, items, on='order_id')
reviews_orders_items.head()
```

```python
print('There are {rows} data in reviews_orders_items'.format(rows=reviews_orders_items.shape[0]))
```

```python
reviews_orders_items.columns
```

### Drop NaN and Unwanted order_status

```python
reviews_orders_items = reviews_orders_items[reviews_orders_items['order_status'].str.match('delivered')]
reviews_orders_items = reviews_orders_items.dropna(axis = 0)
print('There are {rows} data in reviews_orders_items after'.format(rows=reviews_orders_items.shape[0]))
```

### Statistics: product_bought_count, product_average_score

```python
print(reviews_orders_items['review_score'].describe())
```

```python
product_bought_count = reviews_orders_items.groupby(['product_id']).size().reset_index(name='product_bought_count')
print(product_bought_count['product_bought_count'].quantile(np.arange(.9, 1, .01)))
#product_bought_count.head()
```

```python
product_avg_score = reviews_orders_items[['product_id','review_score']].groupby(['product_id'], as_index=False).mean().rename(columns = {'review_score': 'product_avg_score'})
product_avg_score.head()
```

```python
product_statistics = pd.merge(reviews_orders_items[['customer_id','product_id','review_score']], product_bought_count,on='product_id')
product_statistics = pd.merge(product_statistics, product_avg_score, on='product_id')
product_statistics.head()
```

```python
# beacause the mean score is 4.06, therefore we drop the data which score is < 4
score_threshold = 4.0
bought_count_threshold = 6
product_review = product_statistics.query('product_avg_score >= @score_threshold')
product_review = product_review.query('product_bought_count >=@bought_count_threshold')
print('There are {rows} data in reviews_orders_items after'.format(rows=product_review.shape[0]))
```

### New some pivot DataFrames

```python
product_review = product_review.dropna(axis = 0).drop_duplicates(['product_id','customer_id'])
product_review.head(10)
```

```python
product_review['product_avg_score'].astype(np.float16)
print(product_review.shape)
```

```python
customer_product_pivot = product_review.pivot(index = 'product_id',columns='customer_id', values='product_avg_score').fillna(0)
customer_product_pivot
```

### Clustering

```python
import sklearn.neighbors
print(sorted(sklearn.neighbors.VALID_METRICS['brute']))
```

```python
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix

customer_product_matrix = csr_matrix(customer_product_pivot.values) # size is same as pivot table
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
#model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'ball_tree', leaf_size=8, n_neighbors=100)
model_knn.fit(customer_product_matrix)
```

```python
# select a number from (0 ~ number of columns), it repreasents that 'suppose an item is bought, ......'
query_index = np.random.choice(customer_product_pivot.shape[0])

distances, indices = model_knn.kneighbors(customer_product_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors =6)
#distances, indices = model_knn.kneighbors(customer_product_pivot.iloc[1907, :].values.reshape(1, -1), n_neighbors = 6)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(customer_product_pivot.index[query_index]))
        #print('Recommendations for {0}:\n'.format(customer_product_pivot.index[1907]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, customer_product_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```

### SVD

```python
customer_product_pivot2 = product_review.pivot(index = 'customer_id',columns='product_id', values='product_avg_score').fillna(0)
print(customer_product_pivot2.shape)
```

```python
from sklearn.decomposition import TruncatedSVD

X = customer_product_pivot2.values.T
print(X.shape)
SVD = TruncatedSVD(n_components=16,random_state=17)
matrix = SVD.fit_transform(X)
print(matrix.shape)
```

```python
import warnings
warnings.filterwarnings("ignore",category =RuntimeWarning)
corr = np.corrcoef(matrix)
corr.shape
```
