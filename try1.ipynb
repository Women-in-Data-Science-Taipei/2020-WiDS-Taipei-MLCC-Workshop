```python
import pandas as pd
import numpy as np
```

### Read and Merge Data

```python
reviews = pd.read_csv('olist_order_reviews_dataset.csv', encoding='UTF-8')
reviews.head()
```

```python
review_column_drop = ['review_comment_title', 'review_comment_message', 'review_creation_date', 'review_answer_timestamp']
reviews = reviews.drop(review_column_drop, axis=1)
reviews.head()
```

```python
orders = pd.read_csv('olist_orders_dataset.csv', encoding='UTF-8')
orders.head()
```

```python
order_column_drop = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
orders = orders.drop(order_column_drop, axis = 1)
orders.head()
```

```python
items = pd.read_csv('olist_order_items_dataset.csv', encoding='UTF-8')
items.head()
```

```python
item_column_drop = ['shipping_limit_date', 'seller_id']
items = items.drop(item_column_drop, axis=1)
items.head()
```

```python
reviews_orders = pd.merge(reviews,orders, on='order_id')
reviews_orders_items = pd.merge(reviews_orders, items, on='order_id')
reviews_orders_items.head()
```

```python
print('There are {rows} data in reviews_orders_items'.format(rows=reviews_orders_items.shape[0]))
```

```python
reviews_orders_items.columns
```

### Drop NaN and Unwanted order_status

```python
reviews_orders_items = reviews_orders_items[reviews_orders_items['order_status'].str.match('delivered')]
reviews_orders_items = reviews_orders_items.dropna(axis = 0)
print('There are {rows} data in reviews_orders_items after'.format(rows=reviews_orders_items.shape[0]))
```

```python
# make sure there is no NaN value
a = reviews_orders_items.isna()
if False in a:
    print('yes')
```

### Statistics: product_bought_count, product_average_score

```python
print(reviews_orders_items['review_score'].describe())
```

```python
product_bought_count = reviews_orders_items.groupby(['product_id']).size().reset_index(name='product_bought_count')
product_bought_count.head()
```

```python
products = reviews_orders_items[['product_id','review_score']]
product_avg_score = products.groupby(['product_id'], as_index=False).mean()
product_avg_score.head()
```

```python
product_statistics = pd.merge(product_avg_score, product_bought_count,on='product_id')
product_statistics.head()
```

```python
drop_columns = ['review_id', 'order_id', 'review_score', 'order_status','order_item_id','price', 'freight_value']
customer_product = reviews_orders_items.drop(drop_columns, axis=1)
#print(customer_product.columns)
customer_product.head()
```

```python
product_statistics_new = customer_product.merge(product_statistics, left_on='product_id',right_on='product_id')
# product_statistics_new = product_statistics.merge(customer_product,left_on='product_id',right_on='product_id',how='left')
product_statistics_new.head(10)
```

```python
product_statistics_new = customer_product.merge(product_statistics, left_on='product_id',right_on='product_id',how='right')
# product_statistics_new = product_statistics.merge(customer_product,left_on='product_id',right_on='product_id',how='left')
product_statistics_new.head(10)
```
